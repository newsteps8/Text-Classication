```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
import re
import os
import glob
from string import punctuation
import textract
from tqdm import tqdm
import fasttext
```

```python
path = "Dokümanlar/*.pdf"

def pdf_process(path):
    documents = []
    # PDF OKUMA
    for pdf in glob.glob(path): #30 döküman
        text = textract.process(pdf).decode('utf-8').replace("\n", " ").replace("\r", " ")
        #print(text)


        #PREPROCESS OF TEXT
        text = re.sub(r"İ", "i", text)
        text = re.sub(r"/", "",  text)
        text = re.sub(r"  ", " ", text)
        text = re.sub(r"\n", "", text)
        text = re.sub("^\d+\s|\s\d+\s|\s\d+$", " ", text)
        text = ''.join([c for c in text if c not in punctuation])
        text = text.lower()
        documents.append(text)
    return documents
```

```python
model2 = fasttext.train_unsupervised(input="hukuk_docs.csv")


#model2.get_word_vector(model2.words[100])

word_list = model2.words

word_vecs = list()
for word in tqdm(word_list):
	vec = model2.get_word_vector(word)
	word_vecs.append(vec)

#word_vecs can be used for training on model
```

```python
#KMEANS model
true_k = 3 #cluster sayısı
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(word_vecs)

#doc_groups(path)#dökümanları grupla ve keywordlerini çıkart
```
